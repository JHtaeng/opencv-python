```python
img = cv2.imread('C:\git\opencv_python\ch02\lenna.bmp')

img_rotate_90_clockwise = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
cv2.imwrite('lena_cv_rotate_90_clockwise.bmp', img_rotate_90_clockwise)
```

```python
img_rotate_90_clockwise = cv2.cvtColor(img_rotate_90_clockwise, cv2.COLOR_BGR2RGB)
plt.imshow(img_rotate_90_clockwise)
plt.show()
```

```python
img = cv2.imread('C:\git\opencv_python\ch02\lenna.bmp')

img_rotate_90_counterclockwise = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
img_rotate_90_counterclockwise = cv2.cvtColor(img_rotate_90_counterclockwise, cv2.COLOR_BGR2RGB)
plt.imshow(img_rotate_90_counterclockwise)
plt.show()
```

```python
img = cv2.imread('C:\git\opencv_python\ch02\lenna.bmp')

img_rotate_180 = cv2.rotate(img, cv2.ROTATE_180)
img_rotate_180 = cv2.cvtColor(img_rotate_180, cv2.COLOR_BGR2RGB)
plt.imshow(img_rotate_180)
plt.show()
```

- flipcode = 0: flip vertically
- flipcode > 0: flip horizontally
- flipcode < 0: flip vertically and horizontally

```python
img = cv2.imread('C:\git\opencv_python\ch02\lenna.bmp')

img_flip_ud = cv2.flip(img, 0)
img_flip_ud = cv2.cvtColor(img_flip_ud, cv2.COLOR_BGR2RGB)
plt.imshow(img_flip_ud)
plt.show()
```

```python
img = cv2.imread('C:\git\opencv_python\ch02\lenna.bmp')

img_flip_lr = cv2.flip(img, 1)
img_flip_lr = cv2.cvtColor(img_flip_lr, cv2.COLOR_BGR2RGB)
plt.imshow(img_flip_lr)
plt.show()
```

```python
img = cv2.imread('C:\git\opencv_python\ch02\lenna.bmp')

img_flip_ud_lr = cv2.flip(img, -1)
img_flip_ud_lr = cv2.cvtColor(img_flip_ud_lr, cv2.COLOR_BGR2RGB)
plt.imshow(img_flip_ud_lr)
plt.show()
```
