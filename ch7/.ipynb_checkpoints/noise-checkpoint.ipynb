```python
import numpy as np
import cv2
import random
from matplotlib import pyplot as plt
```

```python
def noise_gaussian():
    src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)

    if src is None:
        print('Image load failed!')
        return

    src = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
    plt.imshow(src)
    plt.show()

    for stddev in [10, 20, 30]:
        noise = np.zeros(src.shape, np.int32)
        cv2.randn(noise, 0, stddev)

        dst = cv2.add(src, noise, dtype=cv2.CV_8UC1)

        desc = 'stddev = %d' % stddev
        cv2.putText(dst, desc, (10, 30), cv2.FONT_HERSHEY_SIMPLEX,
                   1.0, 255, 1, cv2.LINE_AA)
        dst = cv2.cvtColor(dst,cv2.COLOR_BGR2RGB)
        plt.imshow(dst)
        plt.show()
```

```python
noise_gaussian()
```

```python
def filter_bilateral():
    src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)

    if src is None:
        print('Image load failed!')
        return

    noise = np.zeros(src.shape, np.int32)
    cv2.randn(noise, 0, 5)
    cv2.add(src, noise, src, dtype=cv2.CV_8UC1)

    dst1 = cv2.GaussianBlur(src, (0, 0), 5)
    dst2 = cv2.bilateralFilter(src, -1, 10, 5)
    
    src = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
    plt.imshow(src)
    plt.show()
    
    dst1 = cv2.cvtColor(dst1,cv2.COLOR_BGR2RGB)
    plt.imshow(dst1)
    plt.show()
    
    dst2 = cv2.cvtColor(dst2,cv2.COLOR_BGR2RGB)
    plt.imshow(dst2)
    plt.show()
```

```python
filter_bilateral()
```

```python
src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)

noise = np.zeros(src.shape, np.int32)
cv2.randn(noise, 0, 5)
cv2.add(src, noise, src, dtype=cv2.CV_8UC1)

dst1 = cv2.bilateralFilter(src, -1, 20, 20)
dst2 = cv2.bilateralFilter(src, -1, 10, 5)

dst1 = cv2.cvtColor(dst1,cv2.COLOR_BGR2RGB)
plt.imshow(dst1)
plt.show()
    
dst2 = cv2.cvtColor(dst2,cv2.COLOR_BGR2RGB)
plt.imshow(dst2)
plt.show()
```

```python
def filter_median():
    src = cv2.imread('lenna.bmp', cv2.IMREAD_GRAYSCALE)

    if src is None:
        print('Image load failed!')
        return

    for i in range(0, int(src.size / 10)):
        x = random.randint(0, src.shape[1] - 1)
        y = random.randint(0, src.shape[0] - 1)
        src[x, y] = (i % 2) * 255

    dst1 = cv2.GaussianBlur(src, (0, 0), 1)
    dst2 = cv2.medianBlur(src, 3)
    
    src = cv2.cvtColor(src,cv2.COLOR_BGR2RGB)
    plt.imshow(src)
    plt.show()
    
    dst1 = cv2.cvtColor(dst1,cv2.COLOR_BGR2RGB)
    plt.imshow(dst1)
    plt.show()
    
    dst2 = cv2.cvtColor(dst2,cv2.COLOR_BGR2RGB)
    plt.imshow(dst2)
    plt.show()
```

```python
filter_median()
```
